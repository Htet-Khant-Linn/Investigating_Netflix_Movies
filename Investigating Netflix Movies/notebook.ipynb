```python
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv('netflix_data.csv')

# print(netflix_df.shape)
for col in netflix_df.columns:
    print(col)
# print(netflix_df.loc[:, 'type'])

netflix_movie = netflix_df[netflix_df['type'] == 'Movie']
# print(netflix_movie)

netflix_movie_1990s = netflix_movie[(netflix_movie['release_year'] >= 1990) & (netflix_movie['release_year'] < 2000)]
print(netflix_movie_1990s)
plt.clf()
plt.hist(netflix_movie_1990s['duration'])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.text(100, 47.5, '100')
plt.show()
plt.clf()

duration = 100

netflix_action_movie_1990s = netflix_movie_1990s[netflix_movie_1990s['genre'] == 'Action']
# print(netflix_action_movie_1990s)
'''
short_movie_count = 0  # Initialize as an integer

for lab, row in netflix_action_movie_1990s.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

print(short_movie_count)
'''
short_movie_count = (netflix_action_movie_1990s['duration'] < 90).sum()
print(short_movie_count)

```
